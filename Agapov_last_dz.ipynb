3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import re
def parse_experience(value):
    
    if value is np.nan or value == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    value_splitted = value.split(' ')[:7]
    years = 0
    months = 0
    for index, item in enumerate (value_splitted):
        if item in year_words:
            years = int(value_splitted[index-1])
        if item in month_words:
            months = int(value_splitted[index-1])
    return int(years*12 + months)

# Преобразование
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(parse_experience)
data.drop(columns=["Опыт работы"], inplace=True)

# Проверка результата
data.head(5)
```

```python
# Вычисление медианного опыта работы (в месяцах) 
median_expirience = data['Опыт работы (месяц)'].median()
print(f"Средний возраст соискателей: {median_expirience}")
```

```python
employment_categories = ['полная занятость', 'частичная занятость', 
                         'проектная работа', 'волонтерство', 'стажировка']

schedule_categories = ['вахтовый метод', 'удаленная работа', 'гибкий график',
                         'полный день', 'сменный график']

for category in employment_categories:
    data[category] = data['Занятость'].str.contains(category).fillna(False)

for category in schedule_categories:
    data[category] = data['График'].str.contains(category).fillna(False)
        
data.drop(columns=['Занятость' , 'График'], inplace=True)

data.head()
```

Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#Загрузим данные о курсе валют
currency_df = pd.read_csv('C:\\Users\\ksaga\\Downloads\\ExchangeRates.csv', sep=',')

currency_df.drop(columns=["per", "time", "vol"], inplace=True)
currency_df.head()
```

```python
# Таблицы соответствий названий валют. ISO-кодировка
currency_mapping = {
    'грн.': 'UAH',
    'USD': 'USD',
    'EUR': 'EUR',
    'руб.':'RUB',
    'белруб.': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}

# Функции для извлечения суммы и валюты

def extract_salary_num(arg):
    return float(arg.split(' ')[0])

def extract_salary_currency(arg):
    curr = arg.split(' ')[1].replace('.', '')
    return currency_mapping.get(curr, None)  # Используем mapping для маппинга валют

# Преобразование данных резюме
data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме'], dayfirst=True).dt.date
data['Сумма зарплаты'] = data['ЗП'].apply(extract_salary_num)
data['Валюта'] = data['ЗП'].apply(extract_salary_currency)
```

```python
# Обработка курсов для RUB (добавление фиктивной строки с пропорцией и курсом 1)
currency_df.loc[currency_df['currency'] == 'RUB', ['proportion', 'close']] = [1, 1]

# Слияние данных резюме и курсов валют
data = data.merge(
    currency_df, 
    left_on=['Валюта', 'Обновление резюме'], 
    right_on=['currency', 'date'],
    how='left'
)


# Заполнение пропусков (для случаев, если курсы валют отсутствуют)
data['proportion'].fillna(1, inplace=True)
data['close'].fillna(1, inplace=True)

# Перевод в рубли
data['ЗП (руб)'] = data['Сумма зарплаты'] * data['close'] / data['proportion']

data.drop(columns=['ЗП', 'Сумма зарплаты', 'Валюта', 'currency', 'date', 'close', 'proportion'], inplace=True)

data.head(3)
```

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
from scipy.stats import zscore

# Логарифмическое преобразование возраста
data['log_Возраст'] = np.log(data['Возраст'])

# Строим распределение в логарифмическом масштабе
plt.figure(figsize=(10, 6))
histplot = sns.histplot(data['log_Возраст'], kde=True)
histplot.axvline(data['log_Возраст'].mean(), color='k', lw=2, label='Среднее')
# Границы интервала трех сигм (с учётом послабления на 4 сигмы в правую сторону)
std_dev = data['log_Возраст'].std()
lower_bound = data['log_Возраст'].mean() - 4 * std_dev
upper_bound = data['log_Возраст'].mean() + 4 * std_dev
histplot.axvline(lower_bound, color='r', linestyle='--', label='Нижняя граница (4 сигмы)')
histplot.axvline(upper_bound, color='r', linestyle='--', label='Верхняя граница (4 сигмы)')

plt.legend()
plt.title("Распределение возраста в логарифмическом масштабе с интервалом 4 сигм")
plt.xlabel("Логарифм возраста")
plt.ylabel("Частота")
plt.show()

# Вычисление Z-оценки для поиска выбросов
data['z_score'] = zscore(data['log_Возраст'])

# Фильтруем выбросы: исключаем значения, которые выходят за пределы 4 сигм в правую сторону
outliers = data[data['z_score'] > 4]

# Выводим выбросы
print(outliers)
```

Распределение возраста в логарифмическом масштабе имеет положительную асимметрию. Это видно по тому, что график имеет более длинный "хвост" в правой части, указывая на присутствие более высоких значений возраста. Линия, обозначающая среднее значение возраста, проходит примерно через центр графика, в районе логарифмического значения возраста около 3,7. Линии, показывающие верхнюю и нижнюю границы интервала в 4 сигмы, проходят довольно далеко от центральной линии среднего. Верхняя граница находится примерно на уровне 4,2, а нижняя - около 3,2 по логарифмической шкале. Выбросы, определенные с использованием метода z-отклонения с пределом в 4 сигмы в правую сторону, будут соответствовать соискателям преклонного возраста. Точное значение возраста, при котором соискатели попадают в категорию выбросов, можно будет определить после построения таблицы с выявленными выбросами.
